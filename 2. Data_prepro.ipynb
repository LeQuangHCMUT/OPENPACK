```python
import pandas as pd

def fill_nan_except_quat(df: pd.DataFrame) -> pd.DataFrame:
    # 1) Xác định cột cần bỏ qua
    quat_columns = [
        col for col in df.columns
        if 'quat' in col.lower() or 'action' in col.lower()
    ]

    for col in df.columns:
        if col in quat_columns:
            continue

        # 2) Bỏ qua cột không phải numeric
        if not pd.api.types.is_numeric_dtype(df[col]):
            continue

        # 3) Interpolate (lấp cả đầu/cuối) rồi fill median
        df[col] = df[col].interpolate(method='linear', limit_direction='both')
        if df[col].isna().any():
            med = df[col].median()
            df[col].fillna(med, inplace=True)

    return df
```

```python
ekf = EKF()

def enforce_unit_quat(q):
    """
    Ép quaternion về dạng đơn vị:
      - Ép kiểu về float64
      - Chuẩn hóa (chia cho norm)
      - Làm tròn đến 12 chữ số thập phân
    """
    q = np.asarray(q, dtype=np.float64)
    norm = np.linalg.norm(q)
    if norm == 0:
        raise ValueError("Quaternion norm is zero.")
    q = q / norm
    return np.around(q, decimals=12)

def calculate_missing_quat(data_file, sensor_prefix):
    # Duyệt từng hàng của DataFrame
    for i in range(len(data_file)):
        # Lấy giá trị cảm biến từ accelerometer và gyroscope
        acc = np.array([
            data_file.at[i, f'{sensor_prefix}/acc_x'],
            data_file.at[i, f'{sensor_prefix}/acc_y'],
            data_file.at[i, f'{sensor_prefix}/acc_z']
        ], dtype=np.float64)
        gyr = np.array([
            data_file.at[i, f'{sensor_prefix}/gyro_x'],
            data_file.at[i, f'{sensor_prefix}/gyro_y'],
            data_file.at[i, f'{sensor_prefix}/gyro_z']
        ], dtype=np.float64)

        # Kiểm tra nếu hàng hiện tại thiếu giá trị quaternion (dựa trên quat_w)
        if pd.isna(data_file.at[i, f'{sensor_prefix}/quat_w']):
            # Nếu là hàng đầu tiên, dùng hàm acc2q để tính quaternion ban đầu
            if i == 0:
                quat = acc2q(acc)
                quat = enforce_unit_quat(quat)
            else:
                # Nếu không, sử dụng quaternion của hàng trước đã tính (đã được đảm bảo đơn vị)
                prev_quat = enforce_unit_quat(np.array([
                    data_file.at[i-1, f'{sensor_prefix}/quat_w'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_x'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_y'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_z']
                ], dtype=np.float64))
                quat = ekf.update(prev_quat, gyr=gyr, acc=acc)
                quat = enforce_unit_quat(quat)

            # Điền giá trị quaternion đã tính vào đúng hàng (in-place)
            data_file.at[i, f'{sensor_prefix}/quat_w'] = quat[0]
            data_file.at[i, f'{sensor_prefix}/quat_x'] = quat[1]
            data_file.at[i, f'{sensor_prefix}/quat_y'] = quat[2]
            data_file.at[i, f'{sensor_prefix}/quat_z'] = quat[3]
        # Nếu hàng có dữ liệu quaternion rồi thì đảm bảo rằng chúng được chuẩn hóa
        else:
            quat = np.array([
                data_file.at[i, f'{sensor_prefix}/quat_w'],
                data_file.at[i, f'{sensor_prefix}/quat_x'],
                data_file.at[i, f'{sensor_prefix}/quat_y'],
                data_file.at[i, f'{sensor_prefix}/quat_z']
            ], dtype=np.float64)
            quat = enforce_unit_quat(quat)
            data_file.at[i, f'{sensor_prefix}/quat_w'] = quat[0]
            data_file.at[i, f'{sensor_prefix}/quat_x'] = quat[1]
            data_file.at[i, f'{sensor_prefix}/quat_y'] = quat[2]
            data_file.at[i, f'{sensor_prefix}/quat_z'] = quat[3]
    return data_file

data_file_2 = pd.read_csv(data_name_file_paths[1])
data_file_3 = pd.read_csv(data_name_file_paths[2])
# Áp dụng cho từng sensor và từng file
for sensor in ['atr01', 'atr02', 'atr03', 'atr04']:
    # for path in data_name_file_paths:
    #     data = pd.read_csv(path)
    #     data = calculate_missing_quat(data, sensor)
    #     data.to_csv(path, index=False)
    data_file_2 = calculate_missing_quat(data_file_2, sensor)
    data_file_3 = calculate_missing_quat(data_file_3, sensor)


```

```python

ekf = EKF()
def enforce_unit_quat(q):
    """
    Ép quaternion về dạng đơn vị:
      - Ép kiểu về float64
      - Chuẩn hóa (chia cho norm)
      - Làm tròn đến 12 chữ số thập phân
    """
    q = np.asarray(q, dtype=np.float64)
    norm = np.linalg.norm(q)
    if norm == 0:
        raise ValueError("Quaternion norm is zero.")
    q = q / norm
    return q

def is_invalid_quat(q):
    q = np.asarray(q, dtype=np.float64)
    if np.any(np.isnan(q)):
        return True
    if np.all(q == 0):
        return True
    return False

def calculate_missing_quat_v2(data_file, sensor_prefix):
    # Duyệt từng hàng của DataFrame
    for i in range(len(data_file)):
        # Lấy các giá trị cảm biến từ accelerometer và gyroscope
        acc = np.array([
            data_file.at[i, f'{sensor_prefix}/acc_x'],
            data_file.at[i, f'{sensor_prefix}/acc_y'],
            data_file.at[i, f'{sensor_prefix}/acc_z']
        ], dtype=np.float64)
        gyr = np.array([
            data_file.at[i, f'{sensor_prefix}/gyro_x'],
            data_file.at[i, f'{sensor_prefix}/gyro_y'],
            data_file.at[i, f'{sensor_prefix}/gyro_z']
        ], dtype=np.float64)

        # Các trường quaternion
        q_fields = [
            f'{sensor_prefix}/quat_w',
            f'{sensor_prefix}/quat_x',
            f'{sensor_prefix}/quat_y',
            f'{sensor_prefix}/quat_z'
        ]
        current_quat = [data_file.at[i, field] for field in q_fields]

        # Nếu bất kỳ phần tử nào của quaternion không hợp lệ (0 hoặc NaN)
        if is_invalid_quat(current_quat):
            # Đặt tất cả các ô quaternion về NaN (hoặc bạn có thể gán 0 nếu cần)
            for field in q_fields:
                data_file.at[i, field] = np.nan

            # Tính lại quaternion dựa vào dữ liệu cảm biến
            if i == 0:
                # Hàng đầu tiên: sử dụng hàm acc2q để khởi tạo
                quat = acc2q(acc)
                quat = enforce_unit_quat(quat)
            else:
                # Hàng sau: sử dụng quaternion của hàng trước (đã hợp lệ) và cập nhật qua EKF
                prev_quat = np.array([
                    data_file.at[i-1, f'{sensor_prefix}/quat_w'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_x'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_y'],
                    data_file.at[i-1, f'{sensor_prefix}/quat_z']
                ], dtype=np.float64)
                prev_quat = enforce_unit_quat(prev_quat)
                quat = ekf.update(prev_quat, gyr=gyr, acc=acc)
                quat = enforce_unit_quat(quat)

            # Gán quaternion đã tính cho DataFrame
            data_file.at[i, f'{sensor_prefix}/quat_w'] = quat[0]
            data_file.at[i, f'{sensor_prefix}/quat_x'] = quat[1]
            data_file.at[i, f'{sensor_prefix}/quat_y'] = quat[2]
            data_file.at[i, f'{sensor_prefix}/quat_z'] = quat[3]
        else:
            # Nếu quaternion hiện có hợp lệ thì chỉ cần chuẩn hóa lại
            quat = np.array(current_quat, dtype=np.float64)
            quat = enforce_unit_quat(quat)
            data_file.at[i, f'{sensor_prefix}/quat_w'] = quat[0]
            data_file.at[i, f'{sensor_prefix}/quat_x'] = quat[1]
            data_file.at[i, f'{sensor_prefix}/quat_y'] = quat[2]
            data_file.at[i, f'{sensor_prefix}/quat_z'] = quat[3]
    return data_file
```
